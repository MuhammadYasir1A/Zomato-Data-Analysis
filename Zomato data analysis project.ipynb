```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe 
```

```python
def handleRate(value):
    value = str(value) .split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe. info()
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(6, 6))  
sns.countplot(data=dataframe, x='listed_in(type)', hue='listed_in(type)', palette="husl", order=dataframe['listed_in(type)'].value_counts().index, legend=False)  
plt.xlabel("Type of Restaurant", fontsize=12, fontweight='bold', color='darkblue')  
plt.ylabel("Count", fontsize=12, fontweight='bold', color='darkblue')  
plt.xticks(rotation=45, ha='right')  
plt.show()  
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()  
plt.plot(grouped_data, c="lime", marker="o", markersize=8, markerfacecolor="yellow")  
plt.xlabel("Type of Restaurant", color="red", fontsize=14)  
plt.ylabel("Votes", color="red", fontsize=14)  
plt.xticks(rotation=45)  
plt.show()  
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=dataframe, x='approx_cost(for two people)', hue='approx_cost(for two people)', palette="viridis", legend=False)
plt.xlabel("Approx Cost (for two people)", fontsize=12, fontweight='bold', color='darkblue')
plt.ylabel("Count", fontsize=12, fontweight='bold', color='darkblue')
plt.xticks(rotation=45)
plt.show()
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe, hue='online_order', palette=["lightcoral", "royalblue"], legend=False)
plt.show()
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(8,6))
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", fmt='d', linewidths=0.5, linecolor='black', cbar_kws={'label': 'Count'})
plt.title("Restaurant Type vs Online Order Heatmap", fontsize=14, fontweight='bold', color='darkred')
plt.xlabel("Online Order Availability", fontsize=12, fontweight='bold', color='darkblue')
plt.ylabel("Type of Restaurant", fontsize=12, fontweight='bold', color='darkblue')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
